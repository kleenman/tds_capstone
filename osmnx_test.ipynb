```python
import pandas as pd
import numpy as np
import osmnx as ox
from pyproj import transform, Proj
```

```python
src_cord = Proj(init='epsg:25832')
dest_cord = Proj(init='epsg:4326')

x = 649000
y = 5487000

lon, lat = transform(src_cord, dest_cord, x, y)

print(f'Latitude: {lat}\nLongitude: {lon}')
```

```python
32685_5339
east = 685000
south = 5339000
west = 684000
north = 5340000

t_east, t_south = transform(src_cord, dest_cord, east, south)
t_west, t_north = transform(src_cord, dest_cord, west, north)

geo_information = ox.features_from_bbox(t_north,t_south,t_east,t_west, tags={'leisure':'park'})
geo_information.head()
```
